### Introduction

#### Reason for analysis
For several years now, I’ve heard acquaintances lament: <i>We don’t have spring anymore.  We skip spring and go straight to summer.  I miss spring-like temperatures.</i>  To see if data supports this belief, I decided to analyze publicly available NOAA weather station data collected from 1910 - 2019 at the Knoxville, TN airport.

#### Definitions
* Spring is March 20 to June 21 (94 days)
* A spring-like temperature is 78℉ or below
* A summer-like temperature is 79℉ or above
* The days of focus are April 1 - May 31, or middle-spring which corresponds to Days 12 - 72 of spring
* The Historical years are the fifteen year span of 1990 - 2004
* The Recent years are the fifteen year span of 2005 - 2019


#### Data source
* URL: https://www.ncdc.noaa.gov/cdo-web/search
* Date Accessed: 7/20/2019
* Parameters for data retrieval: Daily Summaries; 1/1/1910 to 7/20/2019; Station: USW00013891 (Knoxville, TN airport); Custom CSV file;
* Fields in file: Air Temperature (Avg, Max, Min, TOBS)
* All temperatures in degrees Fahrenheit

#### Data Definitions (original data source) : 
* DATE: date of observation
* TAVG: is average temperature for the day
* TMAX: the maximum temperature for the day, aka the "high temp" for the day
* TMIN: the minumumr temperature for the day, aka the "low temp" for the day
* TOBS: is temperature at time of observation
* For more information: https://www.ncdc.noaa.gov/cdo-web/datasets#GHCND 

#### Data Definitions (calculated from original data source)
* Year: {1910-2019}
* MonthDay: {MMDD}
* DayOfSpring: {1-94}
* IsTempTooHotForSpring: {0,1}
* Timeframe: {Recent, Historical}


```r
library(tidyverse)
library(ggplot2)
library(lubridate)
library(gridExtra)
library(grid) 
```

```r
#constants

summer_temp_min <- 78
color_summer_temperature = '#CE6162'
first_year_in_file_expected = 1910
last_year_in_file_expected = 2019
first_year_in_analysis = 1990
first_year_considered_recent = 2005 

#zoom in on the days that seem to be getting hotter and hotter
middle_spring_first_day <- 12   #April 1
middle_spring_last_day <- 72    #May 31
```

```r
GetDataFromFileByMonth <- function(monthInt) {
    in_data <- readr::read_csv("1910_to_present_TempData.csv", 
                               col_types = cols_only (
                                                       'DATE' = col_date(format = ""),
                                                       'TMAX' = col_double()
                                                     )
                              )
    month_data <- in_data %>%
                  filter(as.numeric(strftime(in_data$DATE, "%m")) == monthInt)
}

GetDataFromFileForSpring <- function() {
    in_data <- readr::read_csv("1910_to_2019_Temperatures.csv", 
                               col_types = cols_only (
                                                       'DATE' = col_date(format = ""),
                                                       'TMAX' = col_double()
                                                     )
                              )
    in_data$monthday <- format(in_data$DATE, "%m%d")
    spring_data <- in_data %>%
                  filter(monthday >= "0320" & monthday <= "0621",)
}

GetDayOfSpringBasedOnDate <- function(dt){
    targetYear <- as.numeric(format(dt,format="%Y"))
    firstDayOfSpring <- make_date(year = targetYear, month = 3, day = 20)
    as.numeric(difftime(dt, firstDayOfSpring,units="days")) + 1
}

IsTempTooHotForSpring <- function(temp){
    as.numeric(temp >= summer_temp_min)
}

IsRecent <- function(dt){
    inYear <- as.numeric(format(dt,format="%Y"))
    as.numeric(inYear >= first_year_considered_recent)
}

IsRecentString <- function(dt){
    inYear <- as.numeric(format(dt,format="%Y"))
    IsRecentStringFromYear(inYear)
}

IsRecentStringFromYear <- function(inYear){
    ifelse(inYear >= first_year_considered_recent, "Recent", "Historical")
}
```

```r
spring_data <- GetDataFromFileForSpring()
spring_data$Year <- format(spring_data$DATE,format="%Y")
spring_data$Year <- as.numeric(spring_data$Year)
spring_data$DayOfSpring <- GetDayOfSpringBasedOnDate(spring_data$DATE)
spring_data$IsTempTooHot <- IsTempTooHotForSpring(spring_data$TMAX)

head(spring_data, 1)
```

```r
firstYearInFile <- spring_data$Year[1]
if (firstYearInFile != first_year_in_file_expected) stop ('First row has unexpected Year.  Stopped execution')
```

```r
minHighTemp <- min(spring_data$TMAX)
minHighDay <- spring_data %>%
               filter(TMAX == minHighTemp) %>%
               select(DayOfSpring, Year)
print(minHighTemp)
print(minHighDay)

maxHighTemp <- max(spring_data$TMAX)
maxHighDay <- spring_data %>%
               filter(TMAX == maxHighTemp) %>%
               select(DayOfSpring, Year)
print(maxHighTemp)
print(maxHighDay)
```

```r
ggplot(spring_data, aes(DayOfSpring, Year)) + 
geom_tile(aes(fill = TMAX), colour = "white")  + 
scale_fill_distiller(palette = "Spectral") +
scale_x_continuous(breaks = seq(1, 94, by = 3), expand = c(0, 0)) +
scale_y_continuous(breaks = seq(1911, 2019, by = 2), expand = c(0, 0)) +
labs(
           x = "Day of Spring",
           y = "Year",
           title = "Daily high temperatures generally increase as spring progresses (1910 to 2019)",
           subtitle = "The coldest day (33°F on Day 1) was in 1965 the hottest day (102°F on Day 91) was in 1944",
           fill = "High temperature for the day represented by color"
        ) +
theme (
    axis.text.x = element_text(color = "grey20", size = 7, face = "plain"),
    axis.text.y = element_text(color = "grey20", size = 7, face = "plain"),
    legend.position="bottom"
) 
```

### Analysis of Heatmap
The heatmap above provides an overview of the maximum daily temperatures for spring from 1910 to 2019.  As spring progresses, the temperatures tend to increase.  

```r
spring_data_newest <- spring_data %>%
                      filter (Year >= first_year_in_analysis)
spring_data_newest$IsRecent <- IsRecent(spring_data_newest$DATE)
spring_data_newest$Timeframe <- IsRecentString(spring_data_newest$DATE)
```

```r
if (head(spring_data_newest,1)$Year != first_year_in_analysis) stop ('The Year in the first row is unexpected. Stopped execution')
if (head(spring_data_newest,1)$Timeframe != "Historical") stop ('The Timeframe in the first row is unexpected. Stopped execution')
if (tail(spring_data_newest, 1)$Year != last_year_in_file_expected ) stop ('The Year in the last row is unexpected.  Stopped execution')
```

```r
middle_spring <- spring_data_newest %>%
                 filter (DayOfSpring >= middle_spring_first_day & DayOfSpring <= middle_spring_last_day)
```

```r
if (head(middle_spring,1)$DayOfSpring != middle_spring_first_day) stop ('DayOfSpring in the first row is unexpected')
```

```r
BuildHeatmapTitle <- function(){
    textBlack <- textGrob(expression("Days in middle-spring with " * phantom(bold("summer-like ")) * "temperatures from 1990 to 2019"),
                 vjust=8.5, gp = gpar(col = "black"))
    textEmphasis <- textGrob(expression(phantom("Days in middle-spring with ") * bold("summer-like ") * phantom("temperatures from 1990 to 2019")),
                 vjust=8.5, gp = gpar(col = color_summer_temperature))
    grobTree(textBlack, textEmphasis)
}

summerDaysPlot <- ggplot(middle_spring, aes(DayOfSpring, Year)) + 
geom_tile(aes(fill = IsTempTooHot), colour = "white")  +
scale_fill_gradient(low = '#F5F5F5', high = color_summer_temperature) +
scale_x_continuous(breaks = seq(middle_spring_first_day, middle_spring_last_day, by = 2), expand = c(0, 0)) +
scale_y_continuous(breaks = seq(first_year_in_analysis, 2019, by = 1), expand = c(0, 0)) +
labs (
       x = "Day of Spring",
       y = "Year",
       fill = "High temperature for the day represented by color"
) +
theme (
    axis.text.x = element_text(color = "grey20", size = 7, face = "plain"),
    axis.text.y = element_text(color = "grey20", size = 7, face = "plain"),
    legend.title = element_blank(),
    legend.position = "none"
)  +

coord_equal()

grid.arrange(summerDaysPlot, top = BuildHeatmapTitle())
```

### Analysis of heatmap
The heatmpap above shows middle spring days (April 1 to May 31) with summer-like temperatures from 1990 to 2019. 

<br/>

```r
historical <- middle_spring %>%
              filter (IsRecent == 0)
recent <- middle_spring %>%
          filter(IsRecent == 1)

head(historical, 2)
tail(historical, 2)

head(recent, 2)
tail(recent, 2)
```

```r
if (head(historical, 1)$Year != first_year_in_analysis) stop ("Year in the last row of Historical is unexpected")
if (tail(historical, 1)$Timeframe != "Historical") stop ("Timeframe in the last row of Historical is unexpected")
if (tail(historical, 1)$DayOfSpring != middle_spring_last_day) stop ("DayOfSpring in the last row of Historical is unexpected")

if (head(recent, 1)$Year != first_year_considered_recent) stop ("Year in the first row of Recent is unexpected")
if (tail(recent, 1)$Timeframe != "Recent") stop ("Timeframe in the last row of Recent is unexpected")
if (tail(recent, 1)$DayOfSpring != middle_spring_last_day) stop ("DayOfSpring in the last row of Recent is unexpected")
```

```r
plot1<- ggplot(historical, aes(historical$TMAX)) + geom_histogram()
plot2 <- ggplot(recent, aes(recent$TMAX)) + geom_histogram()
grid.arrange(plot1, plot2, nrow=2)
```

```r
middle_spring_hot_days_by_year <- middle_spring %>%
            filter(IsTempTooHot == 1) %>%
            group_by(Year) %>%
            summarise(NumberOfHotDays=n())

middle_spring_hot_days_by_year$Timeframe <- IsRecentStringFromYear(middle_spring_hot_days_by_year$Year)
```

```r
if (head(middle_spring_hot_days_by_year, 1)$Year != first_year_in_analysis) stop ("Year in the first row of group is unexpected")
if (tail(middle_spring_hot_days_by_year, 1)$Year != last_year_in_file_expected) stop ("Year in the last row of group is unexpected")

head(middle_spring_hot_days_by_year, 35)
summary(middle_spring_hot_days_by_year)
```

```r
ggplot(middle_spring_hot_days_by_year, aes(x=Year, y=NumberOfHotDays)) + 
geom_line() +
geom_point() +
scale_x_continuous(limits = c(first_year_in_analysis, 2019), breaks = seq(first_year_in_analysis, 2019, by = 2) ) +
theme_classic() +
labs (
       x = "Year",
       y = "Number of summer-like days",
       title = "Number of summer-like days in middle-spring (1990 - 2019)",
       subtitle = ""
   ) 
```

### Analysis of plot
The plot above shows the number of summer-like days in middle-spring for each year (1990 - 2019).  It is an alternate view of the previous heatmap.

<br/>

```r
historical <- middle_spring_hot_days_by_year %>%
              filter (Timeframe == 'Historical')

recent <- middle_spring_hot_days_by_year %>%
              filter (Timeframe == 'Recent')


df_summary <- data.frame(Timeframe=character(), Mean=double(), StandardDeviation=double() ,stringsAsFactors=FALSE)

df_summary[nrow(df_summary) + 1,] = list(Timeframe="Historical",
                                        Mean=mean(historical$NumberOfHotDays), 
                                        StandardDeviation = sd(historical$NumberOfHotDays))

df_summary[nrow(df_summary) + 1,] = list(Timeframe="Recent",
                                       Mean=mean(recent$NumberOfHotDays), 
                                       StandardDeviation = sd(recent$NumberOfHotDays))

head(df_summary)
```

```r
ggplot(df_summary, aes(x=Timeframe, y=Mean)) + 
  geom_bar(stat="identity", fill = "#F5F5F5") +
  geom_errorbar(aes(ymin=Mean-StandardDeviation, ymax=Mean+StandardDeviation), width=.2,
                position=position_dodge(.9)) +
   geom_text(aes(label = paste0("Standard Deviation= ", round(StandardDeviation, 4)), 
                 y = Mean-StandardDeviation), vjust = 6.5) +
   labs (
       x = "",
       y = "Mean number of summer days",
       title = "Mean number of summer-like days with standard deviation",
       subtitle = "25 days (Historical) vs 29 days (Recent)"
   ) +
   theme_classic() 
   
```

The plot above shows the mean number of summer days in the Historical group vs the Recent group (25 vs 29 days).  The standard deviation of the Historical group is slightly higher.  The error bars show the values of the mean + or - the standard deviation.  

<br/>

 

```r
ggplot(middle_spring_hot_days_by_year, aes(x=factor(Timeframe), y=NumberOfHotDays)) + 
  geom_boxplot(width=0.6) +
  stat_summary(geom="text", fun.y=quantile,
               aes(label=sprintf("%1.1f", ..y..)),
               position=position_nudge(x=0.40), size=3.5) +
  labs(
          x = "Historical and Recent", 
          y = "Number of summer-like days per year",
          title = "Distributions of the number summer-like days in middle-spring per year",
          subtitle = "In more recent years (2005 - 2019), the number of summer-like days has increased"
        ) +
  theme_classic()
```

### Analysis of plot

The boxplots above show the Historical (1990 - 2004) and Recent (2005 - 2019) distiributions of the number of summer-like days.  The minumum number of summer-like days for Historical vs Recent is 11 vs 21 (including outliers).  The maximum number of summer-like days for Historical vs Recent is 33 vs 38.  Comparing the interquartile range (IQR) for Historical vs Recent, 50% of the values will be in the range of 23-30 days vs 24-33 days.  

In the Historical plot, Quartile 1 is closer to the median than Quartile 3, so the data is not symmetrical.  The data in Quartile 1 is more closely packed than the data in Quartile 3.  In the Recent plot, Quartile 3 is closer to the median than Quartile 1 so the data is not symmetrical.  The data in Quartile 3 is more closely packed than the data in Quartile 1.

By comparing the Historical vs Recent medians (25 vs 31), there is a tendency for the Recent group to have 6 more summer-like days per year.  This does support the perception that there are more summer-like days in middle-spring nowadays.    
 


### Summary
From the methods used, there is evidence to support the perception that summer-like temperatures in middle-spring have happened more often since 2005 in Knoxville, TN.
